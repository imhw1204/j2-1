## class diagram

* class
    * 속성
    * 함수
* visibility

| 접근제어자 | 표기법 | 설명 |
|---|:---:|---|
|public| + |누구나 값을 볼 수 있슴
|private|-|자신만 값을 볼 수 있슴
|protected|#|상속관계인 경우 값을 볼 수 있슴

```python
%%plantuml

@startuml
class Student {
  -stundentNo
  #field1
  +method1()
  +method2()
}
@enduml
```

# 관계
